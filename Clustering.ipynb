```python
from kmodes.kmodes import KModes
import pandas as pd
import numpy as np
```

```python
# suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import xticks
```

```python
# reading the data
data = pd.read_csv("cluster.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
variable_data = data[['Districts','Classification', 'Caste Majority', '1983', '1985', '1989', '1994',
       '1999', 'VM_1999', '2004', 'VM_2004', '2009', 'VM_2009', '2014',
       'VM_2014', '2019', 'VM_2019']]
```

```python
variable_data.describe(include='all')
```

```python
# maintaining copy of categorical data
import copy
training_data = copy.deepcopy(variable_data)
```

```python
# adding indexed labels to data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
variable_data = variable_data.apply(le.fit_transform)
variable_data.head()
```

```python
km_model = KModes(n_clusters=6,init='Huang',n_init=1,verbose=1)
km_clusters = km_model.fit_predict(variable_data)
```

```python
# predicted clusters
km_clusters
```

```python
# sizes of clusters
arr = list(km_clusters)
for i in range(6):
    size = arr.count(i)
    print("Size of cluster-" + str(i) + " = " + str(size))
```

```python
# encoding the data
variable_data = training_data.reset_index()
variable_data.head()
```

```python
# combining the data
clusters_data = pd.DataFrame(km_clusters)
clusters_data.columns = ['predicted_cluster']
combined_data = pd.concat([variable_data, clusters_data],axis=1).reset_index()
combined_data = combined_data.drop(['index','level_0'],axis=1)
```

```python
labels = data[['Constituency_Number','Constituency_Name']]
combined_data_with_labels = pd.concat([labels,combined_data],axis=1).reset_index()
combined_data_with_labels = combined_data_with_labels.drop(['index'],axis=1)
combined_data_with_labels.head()
```
